**Create a Dataframe**

```python
import pandas as pd
```

```python
# Create a DataFrame
data = {'Season': ['summer', 'autumn', 'spring', 'winter', 'autumn']}
df = pd.DataFrame(data)
df.head()
```

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df['season label']= le.fit_transform(df['Season'])
df.head()
```

```python
# Customize/ Map the label

mapping= { 'spring':0, 'summer':1, 'autumn':2, 'winter':3}
df['season_custom_label']= df['Season'].map(mapping)
df.head()
```
