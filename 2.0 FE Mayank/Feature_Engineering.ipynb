```python
numerical_features_nan= [feature for feature in df.columns if df[feature].dtypes!='object' and
                           df[feature].isnull().sum()>=1]
```

```python
numerical_features_nan
```

```python
for feature in numerical_features_nan:
  df[feature].plot(kind='box', title='Outlier in column '+ feature)
  plt.show()
```

```python
for feature in numerical_features_nan:
  # replace outliers with median
  median_values= df[feature].median()
  # to capture nan create new feature
  df[feature+ 'nan']= np.where(df[feature].isnull(),1,0)
  df[feature].fillna(median_values, inplace=True)
```

```python
#Temporal Variables

for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
  df[feature + 'age of house']= df['YrSold']- df[feature]
```
