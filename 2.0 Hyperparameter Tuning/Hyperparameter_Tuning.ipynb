**Import Dependencies**

```python
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

**Import Dataset**

```python
breast_cancer= sklearn.datasets.load_breast_cancer()
```

```python
# loading data to a dataframe
df= pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
# Adding target colum to dataframe
df['label']= breast_cancer.target
```

```python
df.head()
```

```python
df.columns
```

```python
df['label'].value_counts()
```

1-----> Benign

0-----> Malignant

**Features & Target**

```python
y= df['label']
X= df.drop(columns='label', axis=1)
```

```python
X=np.asarray(X)
y=np.asarray(y)
```

##**GridSearch CV**

It is used to determine the best parameters for our model

```python
# loading model
model= SVC()
```

```python
# Hyperparameters

parameters={
    'kernel':['linear', 'poly','rbf','sigmoid'],
    'C':[1,5,10,20]

}
```

```python
# Grid Search

classifier= GridSearchCV(model, parameters, cv=5)
```

```python
# Fit the data to the Model

classifier.fit(X,y)
```

```python
# Best Parameters


best_parameters= classifier.best_params_
print(best_parameters)
```

```python
# Highest Accuracy

highest_accuracy= classifier.best_score_
print(highest_accuracy)
```

```python
result= pd.DataFrame(classifier.cv_results_)
result.head()
```

```python
grid_search_result= result[['param_C', 'param_kernel','mean_test_score']]
grid_search_result
```

Highest Accuracy: 95.2%

Best Parameters:  'C':10,  Kernel: Linear

##**Randomized Search CV**

```python
# loading model
model= SVC()
```

```python
# Hyperparameters

parameters={
    'kernel':['linear', 'poly','rbf','sigmoid'],
    'C':[1,5,10,20]

}
```

```python
# Randomised Search CV

classifier= RandomizedSearchCV(model, parameters, cv=5)
```

```python
# Fit the data to the Model

classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
result= pd.DataFrame(classifier.cv_results_)
result.head()
```

```python
# Best Parameters


best_parameters= classifier.best_params_
print(best_parameters)
```

```python
# Highest Accuracy

highest_accuracy= classifier.best_score_
print(highest_accuracy)
```

```python
randomized_search_result= result[['param_C', 'param_kernel','mean_test_score']]
randomized_search_result
```

Highest Accuracy: 95.2%

Best Parameters: 'C':10, Kernel: Linear
